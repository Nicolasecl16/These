# Experience 1 :
On crée un jeu de données distribué selon une loi normale multivariée. 
Ensuite, on crée une matrice donnant les paramètres de forme dans l'espace. 
Enfin, on transforme les marginales en GEV

```python
import numpy as np
from scipy import interpolate
from scipy.stats import norm

N = 10
a = np.random.randint(0,5,(100,100))
S=np.matmul(a,np.transpose(a))
print(S/100)
```

```python
N_grid = N
M = np.eye(N_grid*N_grid)+ S/(N_grid*N_grid)
print(M)
D = np.random.multivariate_normal(np.zeros(N_grid*N_grid), M, size = 10000)
print(D[:3,:])
D=D.reshape(10000,10,10)
print(D[:3,:,:])

```

```python
xi_mat=np.zeros((N_grid,N_grid))
x = np.random.randint(0,N_grid,10)
y=np.random.randint(0,N_grid,10)
z = np.random.randint(0,2,10)
fun = interpolate.interp2d(x, y, z, kind='linear')
for i in range(N_grid):
    for j in range(N_grid):
        xi_mat[i,j] = fun(i,j)
```

```python
Ind = np.where(xi_mat<0)
xi_mat[Ind]=0
Ind2 = np.where(xi_mat>2)
xi_mat[Ind2]=2
print(xi_mat)
```

```python
Data_Unif =np.zeros(D.shape)
print(Data_Unif.shape)

for i in range(N_grid):
    for j in range(N_grid):
        Data_Unif[:,i,j] = norm.cdf(D[:,i,j]/(np.diag(M)[N_grid*i+j]**2))
    
print(Data_Unif)
```

```python
def inv_quantile(p,ksi):
    if ksi == 0:
        return -np.log(-np.log(p))
    else :
        return 1/ksi*((-np.log(p))**(-ksi)-1)
        
```

```python
Data_tot =np.zeros(D.shape)
for i in range(N_grid):
    for j in range(N_grid):
        Data_tot[:,i,j] = inv_quantile(Data_Unif[:,i,j],xi_mat[i,j])
```

```python
import matplotlib.pyplot as plt
import torch as T

class ABCnet(T.nn.Module):
    def __init__(self):
        super(ABCnet, self).__init__()
        self.c1 = T.nn.Linear(4, 20)
        self.c2 = T.nn.Linear(20,40)
        self.c3 = T.nn.Linear(40,20)
        self.c4 = T.nn.Linear(20,1)
        
        self.c5 = T.nn.Linear(4, 20)
        self.c6 = T.nn.Linear(20,40)
        self.c7 = T.nn.Linear(40,20)
        self.c8 = T.nn.Linear(20,1)
        
        self.c9 = T.nn.Linear(4, 20)
        self.c10 = T.nn.Linear(20,40)
        self.c11 = T.nn.Linear(40,20)
        self.c12 = T.nn.Linear(20,1)
        
        T.nn.init.zeros_(self.c5.weight)
        T.nn.init.zeros_(self.c5.bias)
        T.nn.init.zeros_(self.c6.weight)
        T.nn.init.zeros_(self.c6.bias)
        T.nn.init.zeros_(self.c7.weight)
        T.nn.init.zeros_(self.c7.bias)
        T.nn.init.zeros_(self.c8.weight)
        T.nn.init.zeros_(self.c8.bias)
        
        T.nn.init.zeros_(self.c5.weight)
        T.nn.init.constant_(self.c5.bias,1)
        T.nn.init.zeros_(self.c6.weight)
        T.nn.init.constant_(self.c6.bias,1)
        T.nn.init.zeros_(self.c7.weight)
        T.nn.init.constant_(self.c7.bias,1)
        T.nn.init.zeros_(self.c8.weight)
        T.nn.init.constant_(self.c8.bias,1)
        
        T.nn.init.zeros_(self.c9.weight)
        T.nn.init.constant_(self.c9.bias,1)
        T.nn.init.zeros_(self.c10.weight)
        T.nn.init.constant_(self.c10.bias,1)
        T.nn.init.zeros_(self.c11.weight)
        T.nn.init.constant_(self.c11.bias,1)
        T.nn.init.zeros_(self.c12.weight)
        T.nn.init.constant_(self.c12.bias,1)
        
    def a(self, x):
        z = T.tanh(self.c1(x))
        z = T.tanh(self.c2(z))
        z = T.tanh(self.c3(z))
        z = self.c4(z)
        z=T.unsqueeze(z,dim=0)
        return z  

    def b(self, x):
        z = T.tanh(self.c5(x))
        z = T.tanh(self.c6(z))
        z = T.tanh(self.c7(z))
        z = self.c8(z)
        z=T.unsqueeze(z,dim=0)
        return z  
    
    def c(self, x):
        z = T.tanh(self.c9(x))
        z = T.tanh(self.c10(z))
        z = T.tanh(self.c11(z))
        z = self.c12(z)
        z=T.unsqueeze(z,dim=0)
        return T.abs(z)  

    def forward(self, x):
        z = T.cat((self.a(x),self.b(x),self.c(x)),0)
        return z

Res = ABCnet()
```

```python
class coords():
    x =T.arange(0,10)
    y = T.arange(0,10)
    xy = T.meshgrid(x,y)
c=coords()
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
from scipy.misc import derivative

F = []
for i in range(N_grid):
    for j in range(N_grid):
        F.append(ECDF(Data_tot[:,i,j]))
F=np.asarray(F)       
F = F.reshape(N_grid,N_grid)
#F=T.Tensor(F)

def D(f,x):
    return(derivative(f,x,dx=1e-2))

def DF(F,A):
    S=T.shape(A)
    res=T.zeros((S[0],S[1]))
    for i in range(S[0]):
        for j in rage(S[1]):
            res[i,j] = (D(F[i,j],A[i,j]))
    return res


```

```python
co = T.cat((T.unsqueeze(c.xy[0],dim=-1),T.unsqueeze(c.xy[1],dim=-1)),dim=2)
```

```python
## A adapter pour Res1 = T.Tensor([F[i,j](Mat_h) for i in range(N_grid) for j in range(N_grid)])

class LegendrePolynomial3(T.autograd.Function):
    """
    We can implement our own custom autograd Functions by subclassing
    torch.autograd.Function and implementing the forward and backward passes
    which operate on Tensors.
    """

    @staticmethod
    def forward(ctx, input):
        """
        In the forward pass we receive a Tensor containing the input and return
        a Tensor containing the output. ctx is a context object that can be used
        to stash information for backward computation. You can cache arbitrary
        objects for use in the backward pass using the ctx.save_for_backward method.
        """
        ctx.save_for_backward(input)
        return 0.5 * (5 * input ** 3 - 3 * input)

    @staticmethod
    def backward(ctx, grad_output):
        """
        In the backward pass we receive a Tensor containing the gradient of the loss
        with respect to the output, and we need to compute the gradient of the loss
        with respect to the input.
        """
        input, = ctx.saved_tensors
        return grad_output * 1.5 * (5 * input ** 2 - 1)
    
class Auto_F(T.autograd.Function):
    @staticmethod
    def forward(ctx,input):
        ctx.save_for_backward(input)
        input1,input2, F, F2, F_d, N_grid = input
        Res1 = T.Tensor([F2[i,j](input1) for i in range(N_grid) for j in range(N_grid)])
        Res2 = T.Tensor([F[i,j](input2) for i in range(N_grid) for j in range(N_grid)])
        return Res2-Res1
    
    @staticmethod
    def backward(ctx, grad_output):
        input, = ctx.saved_tensors
        input1,input2, F, F_d, N_grid = input
        G1 = F_d(input1)
        G2 = F_d(input2) 
        print(grad_output)
        return grad_output*(G1,-G2)
       
```

```python
def h(x,a,b,c):
    return a+b*x**c
    
def h_inv(x,a,b,c):
    return ((x-a)/b)**(1/c)



def loss_funct(F,x,val,val2):
    Mat_h = h(val2,x[0,:,:],x[1,:,:],x[2,:,:])
    Mat_h_inv = h_inv(val,x[0,:,:],x[1,:,:],x[2,:,:])
    Res1 = T.Tensor([F[i,j](Mat_h) for i in range(N_grid) for j in range(N_grid)])
    Res2 = [F[i,j](Mat_h_inv) for i in range(N_grid) for j in range(N_grid)]
    Res1 = T.Tensor(Res1)
    Res2 = T.Tensor(Res2)
    Res_Tot = Res2-Res1
    return T.mean(Res_Tot**2)

def loss_funct2(F,F2,F_d,x,val,val2):
    Mat_h = h(val2,0*x[0,:,:],0*x[1,:,:],x[2,:,:])
    Mat_h_inv = h_inv(val,x[0,:,:],x[1,:,:],x[2,:,:])
    Res_tot = Auto_F.apply((Mat_h, Mat_h_inv, F,F2 ,F_d ,N_grid))
    return T.mean(Res_tot**2)

    
```

```python
# 0. setup
print("\nBegin training")
T.manual_seed(1)
np.random.seed(1)

# 1. create DataLoader object
print("\nImporting dataset ")

bat_size = 50
train_ldr = T.utils.data.DataLoader(Data_tot,
batch_size=bat_size, shuffle=True)

# 2. train model
max_epochs = 1000
ep_log_interval = 1
lrn_rate = 0.001

loss_func = loss_funct2
optimizer = T.optim.Adam(Res.parameters(), lr=lrn_rate)

print("\nbat_size = %3d " % bat_size)
print("loss = " + str(loss_func))
print("optimizer = Adam")
print("max_epochs = %3d " % max_epochs)
print("lrn_rate = %0.3f " % lrn_rate)

print("\nStarting training")
Res = Res.train()
for epoch in range(0, max_epochs):
    epoch_loss = 0  # for one full epoch

    for (batch_idx, batch) in enumerate(train_ldr):
        X = batch 
        #creating randomly permute coordinate
        x2 = T.randperm(N_grid)
        y2 = T.randperm(N_grid)
        xy2 = T.meshgrid(x2,y2)
        co2 = T.cat((T.unsqueeze(xy2[0],dim=-1),T.unsqueeze(xy2[1],dim=-1)),dim=2)
        X2 = T.transpose(X,0,1)
        X2 = T.transpose(X2,2,1)
        X2 = X2[xy2]
        X2 = T.transpose(X2,2,1)
        X2 = T.transpose(X2,0,1)
        F2 = F[xy2]
        optimizer.zero_grad()
        outpt = Res(T.cat((co.float(),co2.float()),dim=2))
        outpt= outpt.reshape(3,N_grid,N_grid)
        loss_obj = loss_func(F, F2, DF, outpt, X,X2)
        epoch_loss += loss_obj.item() 
        loss_obj.requires_grad_(True)
        loss_obj.backward()
        optimizer.step()

    if epoch % ep_log_interval == 0:
        print("epoch = %4d   loss = %0.4f" % (epoch, epoch_loss))

print("Done ")

```
