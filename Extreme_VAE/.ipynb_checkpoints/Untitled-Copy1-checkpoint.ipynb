## Data generation

Experimentation : Sabourin PCA, Segers, J. (2012). Max-stable models for multivariate extremes



```python
import numpy as np
from scipy.stats import weibull_max
import matplotlib.pyplot as plt
import torch as T

d = 100
subspace = 3
N_set = 10000
alphas = np.random.uniform(low=1, high=10, size=subspace)
print(alphas)

dataset = np.zeros((N_set,d))

for i in range(subspace):
        dataset[:,i] = np.random.gamma(alphas[i], scale=1.0, size=N_set)
```

```python
d_sum = np.sum(dataset,1)

for i in range(subspace):
    dataset[:,i]= dataset[:,i]/d_sum
print(np.max(dataset))
r = weibull_max.rvs(2, size=10000)
print(np.min(r))
for i in range(subspace):
    dataset[:,i]=dataset[:,i]*(-r)
print(np.min(dataset))
print(np.max(dataset))

    
```

```python
Sig = np.zeros((100,100))
Sig+=0.2
Sig = Sig-10/d*np.identity(100)

noise =  0.1*np.random.multivariate_normal(mean=np.zeros(100), cov=Sig, size=N_set)
print(noise[1,:30])
```

```python
Data_tot = noise+dataset
print(Data_tot[1,:30])
norm = np.sqrt(np.sum(Data_tot**2,1))
print(norm)
u = np.quantile(norm,0.9)
print(u)
Ind =np.where(norm>u)
print(Ind[0])
print(Data_tot[Ind[0],:20])
dataset_EVT = np.zeros((1000,100))
print(np.shape(1/norm[Ind]))
for i in range(100):
    dataset_EVT[:,i] = 1/norm[Ind[0]]*Data_tot[Ind[0],i]
print(dataset_EVT)
```

```python
dataset_EVT =np.float32(dataset_EVT)
dataset_EVT=T.FloatTensor(dataset_EVT)
```

## Auto-encoder model

```python
device = T.device("cpu")  # apply to Tensor or Module

class AE(T.nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        self.enc1 = T.nn.Linear(100, 50)  
        self.enc2 = T.nn.Linear(50, 2)
        self.dec1 = T.nn.Linear(2, 50)
        self.dec2 = T.nn.Linear(50, 100)

        T.nn.init.xavier_uniform_(self.enc1.weight)
        T.nn.init.zeros_(self.enc1.bias)
        T.nn.init.xavier_uniform_(self.enc2.weight)
        T.nn.init.zeros_(self.enc2.bias)
        T.nn.init.xavier_uniform_(self.dec1.weight)
        T.nn.init.zeros_(self.dec1.bias)
        T.nn.init.xavier_uniform_(self.dec2.weight)
        T.nn.init.zeros_(self.dec2.bias)

    def encode(self, x):
        z = T.tanh(self.enc1(x))
        z = T.tanh(self.enc2(z))  # act depends on scenario
        return z  

    def decode(self, x):
        z = T.tanh(self.dec1(x))
        z = self.dec2(z)  # no activation
        return z        

    def forward(self, x):
        z = self.encode(x)
        z = self.decode(z)
        return z
```

```python
# 0. setup
print("\nBegin AE training")
T.manual_seed(1)
np.random.seed(1)

# 1. create DataLoader object
print("\nImporting dataset ")

  # all 1797 rows

bat_size = 10
train_ldr = T.utils.data.DataLoader(dataset_EVT[0:800],
batch_size=bat_size, shuffle=True)

# 2. create network
print("\nCreating 100-50-10-50-100 autoencoder ")
net = AE().to(device)

# 3. train model
max_epochs = 150
ep_log_interval = 10
lrn_rate = 0.001

loss_func = T.nn.MSELoss() 
optimizer = T.optim.Adam(net.parameters(), lr=lrn_rate)

print("\nbat_size = %3d " % bat_size)
print("loss = " + str(loss_func))
print("optimizer = Adam")
print("max_epochs = %3d " % max_epochs)
print("lrn_rate = %0.3f " % lrn_rate)

print("\nStarting training")
net = net.train()
for epoch in range(0, max_epochs):
    epoch_loss = 0  # for one full epoch

    for (batch_idx, batch) in enumerate(train_ldr):
        X = batch # no targets needed

        optimizer.zero_grad()
        oupt = net(X)
        loss_obj = loss_func(oupt, X)+(1-T.mean(T.sqrt(T.sum(oupt**2,1))))**2  # note: X not Y
        epoch_loss += loss_obj.item() #+(1-T.mean(T.sqrt(T.sum(oupt**2,1))))**2 # accumulate
        loss_obj.backward()
        optimizer.step()

    if epoch % ep_log_interval == 0:
        print("epoch = %4d   loss = %0.4f" % (epoch, epoch_loss))
        Validation_set = dataset_EVT[800:]
        out = net(Validation_set)
        N=T.sqrt(T.sum(out**2,1))
        new_out =T.zeros(200,100)
        for i in range(200):
            new_out[i] = out[i,:]/N[i]
        print(T.sum((new_out-dataset_EVT[800:])**2)/200)
print("Done ")

```

# Variational auto-encoder

```python
import matplotlib.pyplot as plt # plotting library
import numpy as np # this module is useful to work with numerical arrays
import pandas as pd 
import random 
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader,random_split
from torch import nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class VariationalEncoder(nn.Module):
    def __init__(self, latent_dims):  
        super(VariationalEncoder, self).__init__()
        self.enc1 = T.nn.Linear(100, 50)  
        self.enc2 = T.nn.Linear(50, 25)
        self.linear2 = nn.Linear(25, latent_dims)
        self.linear3 = nn.Linear(25, latent_dims)

        self.N = torch.distributions.Normal(0, 1)
        self.N.loc = self.N.loc.cuda() # hack to get sampling on the GPU
        self.N.scale = self.N.scale.cuda()
        self.kl = 0

    def forward(self, x):
        x = x.to(device)
        x = F.relu(self.enc1(x))
        x = F.relu(self.enc2(x))
        mu =  self.linear2(x)
        sigma = torch.exp(self.linear3(x))
        z = mu + sigma*self.N.sample(mu.shape)
        self.kl = (sigma**2 + mu**2 - torch.log(sigma) - 1/2).sum()
        return z   
```

```python
class Decoder(nn.Module):
    
    def __init__(self, latent_dims):
        super().__init__()

        self.decoder_lin = nn.Sequential(
            nn.Linear(latent_dims, 25),
            nn.ReLU(True),
            nn.Linear(25, 50),
            nn.ReLU(True),
            nn.Linear(50, 100),
        )
        
    def forward(self, x):
        print(x.get_device())
        x = self.decoder_lin(x)
        return x
```

```python
class VariationalAutoencoder(nn.Module):
    def __init__(self, latent_dims):
        super(VariationalAutoencoder, self).__init__()
        self.encoder = VariationalEncoder(latent_dims)
        self.decoder = Decoder(latent_dims)

    def forward(self, x):
        x = x.to(device)
        z = self.encoder(x)
        return self.decoder(z)
```

```python
### Set the random seed for reproducible results
torch.manual_seed(0)

d = 2

vae = VariationalAutoencoder(latent_dims=d)

lr = 1e-4

optim = torch.optim.Adam(vae.parameters(), lr=lr, weight_decay=1e-5)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(f'Selected device: {device}')

vae.to(device)
```

```python
### Training function
def train_epoch(vae, device, dataloader, optimizer):
    # Set train mode for both the encoder and the decoder
    vae.train()
    train_loss = 0.0
    # Iterate the dataloader (we do not need the label values, this is unsupervised learning)
    for x in dataloader: 
        # Move tensor to the proper device
        x = x.to(device)
        x_hat = vae(x)
        # Evaluate loss
        loss = 0.1*((x - x_hat)**2).sum() + 0.001*vae.encoder.kl + (1-T.mean(T.sqrt(T.sum(x_hat**2,1))))**2

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        # Print batch loss
        print('\t partial train loss (single batch): %f' % (loss.item()))
        train_loss+=loss.item()

    return train_loss / len(dataloader.dataset)
```

```python
num_epochs = 50

for epoch in range(num_epochs):
    train_loss = train_epoch(vae,device,train_ldr,optim)
    #val_loss = test_epoch(vae,device,valid_loader)
    #print('\n EPOCH {}/{} \t train loss {:.3f} \t val loss {:.3f}'.format(epoch + 1, num_epochs,train_loss,val_loss))
    #plot_ae_outputs(vae.encoder,vae.decoder,n=10)
```

### Plotting Dirichlet PDF

```python
import math

def Diri_pdf(x,alphas):
    num = 0
    den =1
    var = 1
    n=len(alphas)
    for i in range(n):
        alpha_i = alphas[i]
        num+=alpha_i
        den = den*math.gamma(alpha_i)
        var = var*(x[i]**(alpha_i-1))                   
    return (den/num*var)
```

```python
import matplotlib.tri as tri
corners = np.array([[0, 0], [1, 0], [0.5, 0.75**0.5]])
AREA = 0.5 * 1 * 0.75**0.5
triangle = tri.Triangulation(corners[:, 0], corners[:, 1])

refiner = tri.UniformTriRefiner(triangle)
trimesh = refiner.refine_triangulation(subdiv=4)

plt.figure(figsize=(8, 4))
for (i, mesh) in enumerate((triangle, trimesh)):
    plt.subplot(1, 2, i+ 1)
    plt.triplot(mesh)
    plt.axis('off')
    plt.axis('equal')
```

```python
# For each corner of the triangle, the pair of other corners
pairs = [corners[np.roll(range(3), -i)[1:]] for i in range(3)]
# The area of the triangle formed by point xy and another pair or points
tri_area = lambda xy, pair: 0.5 * np.linalg.norm(np.cross(*(pair - xy)))

def xy2bc(xy, tol=1.e-4):
    '''Converts 2D Cartesian coordinates to barycentric.'''
    coords = np.array([tri_area(xy, p) for p in pairs]) / AREA
    return np.clip(coords, tol, 1.0 - tol)
```

```python
class Dirichlet(object):
    def __init__(self, alpha):
        from math import gamma
        from operator import mul
        self._alpha = np.array(alpha)
        self._coef = gamma(np.sum(self._alpha)) / \
                           np.multiply.reduce([gamma(a) for a in self._alpha])
    def pdf(self, x):
        '''Returns pdf value for `x`.'''
        from operator import mul
        return self._coef * np.multiply.reduce([xx ** (aa - 1)
                                               for (xx, aa)in zip(x, self._alpha)])
```

```python
def draw_pdf_contours(dist, nlevels=200, subdiv=8, **kwargs):
    import math

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv=subdiv)
    num = [xy2bc(xy) for xy in zip(trimesh.x, trimesh.y)]
    pvals = [dist.pdf(xy2bc(xy)) for xy in zip(trimesh.x, trimesh.y)]
    plt.tricontourf(trimesh, pvals, nlevels, cmap='jet', **kwargs)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
```
